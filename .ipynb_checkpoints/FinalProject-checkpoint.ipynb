### Models (Logistic, SVM, Decision Tree and Random Forecast

#### Decision Tree

```python
df4 = df2
df3 = df2.drop(columns = ['fnlwgt'])
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df3.drop('salary',axis=1), 
                                                    df3['salary'], test_size=0.20, 
                                                    random_state=101)
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=10, criterion='entropy')
model.fit(X_train,Y_train)

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn import metrics
Y_pred = model.predict(X_test)
predictions = model.predict(X_test)
print(confusion_matrix(Y_test,predictions))
print(classification_report(Y_test,predictions))
print(accuracy_score(Y_test, predictions))
```
